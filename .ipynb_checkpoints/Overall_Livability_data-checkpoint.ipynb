```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import math

Zhonghe_real_estate = pd.read_csv('./data/real_estate_cleaned.csv', encoding = 'utf-8')
print(Zhonghe_real_estate)

# 將圖表標題改為標楷體
import matplotlib
matplotlib.rc('font', family = 'DFKai-SB')

# 提高jupyter notebook圖形顥示解析度
%config InlineBackend.figure_format = 'retina'

# 使用boxplot觀察'單價元平方公尺'在不同'分類標籤'間的分佈差異。
ax = sns.boxplot(x = '分類標籤', y = '單價元平方公尺', data = Zhonghe_real_estate)

# 調整seaborn圖表大小
ax.figure.set_size_inches(18, 9)
plt.title('觀察"單價元平方公尺"在不同道路間的分佈差異', fontsize = 20)

# 將x軸標籤旋90度
plt.setp(ax.get_xticklabels(), rotation = 90, fontsize = 12)
plt.ylabel('單價元平方公尺')
plt.show()

# 依不同道路標籤繪製直方圖
fig = plt.figure()
fig.figure.set_size_inches(12, 24)
for road in set(Zhonghe_real_estate['分類標籤']):
    dfroad = Zhonghe_real_estate[Zhonghe_real_estate['分類標籤'] == road]
    dfroad['單價元平方公尺'][dfroad['單價元平方公尺'] < 400000].hist(alpha = 0.7)
plt.xlim(0, 400000)
plt.legend(set(Zhonghe_real_estate['分類標籤']),loc = 0, bbox_to_anchor = (1.05, 1.0), borderaxespad = 0.)
plt.title('依"道路名稱"畫出"單價元平方公尺"之價格分佈圖', fontsize = 30)

plt.show()
```

```python
# 將'距捷運站距離'、'學校距離'、'超市距離'、'公園距離'加入GeoPandas中

# 讀取中和區範圍的shp
county_data = gpd.read_file('./圖像資料/鄉(鎮、市、區)界線(TWD97經緯度)1140318/TOWN_MOI_1140318.shp')
print(county_data.head())
Zhonghe_District_data = county_data[county_data['TOWNNAME'] == '中和區']

# 將房屋轉成GeoDataFrame
df = pd.read_csv('./data/real_estate_cleaned.csv', encoding = 'utf-8')

gdf_real_estate = gpd.GeoDataFrame(df, geometry = gpd.points_from_xy(df['Lng'], df['Lat']), crs = 'EPSG: 4326')
gdf_real_estate = gdf_real_estate.to_crs(Zhonghe_District_data.crs)

# 計算距離最近捷運站的距離(先轉成平面座標)

gdf_real_estate = gdf_real_estate.to_crs(epsg = 3826)

df_mrt = pd.read_csv('./data/northern-taiwan-mrt-station.csv', encoding = 'utf-8')
gdf_mrt = gpd.GeoDataFrame(df_mrt, geometry = gpd.points_from_xy(df_mrt['lon'], df_mrt['lat']), crs = 'EPSG: 4326')
gdf_mrt = gdf_mrt.to_crs(epsg = 3826)
Zhonghe_gdf_mrt1 = gdf_mrt[gdf_mrt['address'].str.contains('新北市中和區', na = False)]

# 去除'景安'站的重複值
Zhonghe_gdf_mrt = Zhonghe_gdf_mrt1.drop_duplicates(subset = ['station_name_tw'])

print(Zhonghe_gdf_mrt)
print(Zhonghe_gdf_mrt.shape)

gdf_real_estate['距捷運站距離'] = gdf_real_estate.geometry.apply(lambda p: Zhonghe_gdf_mrt.distance(p).min())

gdf_real_estate['距離分類(捷運站)'] = pd.cut(gdf_real_estate['距捷運站距離'],
                                bins = [0, 500, 1000, 1500, 2000],
                                labels = ['0-500m', '500-1000m', '1000-1500m', '1500-2000m'])
print(gdf_real_estate)

# 統計各種不同捷運站距離的房屋數量
mrt_distance = gdf_real_estate.groupby('距離分類(捷運站)')
print(mrt_distance.size())

# 設定捷運站緩衝區
Zhonghe_gdf_mrt['buffer_500m(subway_station)'] = Zhonghe_gdf_mrt.buffer(500)

# 建立站點房價列表
result_list_subway_station = []

for idx, station in Zhonghe_gdf_mrt.iterrows():
    nearby = gdf_real_estate[gdf_real_estate.geometry.within(station['buffer_500m(subway_station)'])]
    mean_price = nearby['單價元平方公尺'].mean()
    result_list_subway_station.append({'站名': station['station_name_tw'], '平均房價': mean_price, '房屋數量': len(nearby)})
df_station_prices = pd.DataFrame(result_list_subway_station)
print(result_list_subway_station)

print(df_station_prices)

# df_station_prices.to_csv('./data/捷運站房價統計表.csv', index = 0, header = 1, encoding = 'utf-8')

# 算出各種不同捷運站距離的房屋均價
print(mrt_distance['單價元平方公尺'].mean())

# 繪出'距捷運站距離'直方圖
sns.histplot(gdf_real_estate['距離分類(捷運站)'], kde = False)
plt.show()

# 用boxplot觀察'單價元平方公尺'在不同'距離分類(捷運站)'區間的分佈差異
a = gdf_real_estate.sort_values(by = '距離分類(捷運站)', ascending = True)
ax = sns.boxplot(x = '距離分類(捷運站)', y = '單價元平方公尺', data = a)
plt.xlabel('距離分類(捷運站)', fontsize = 18)
plt.ylabel('單價元平方公尺', fontsize = 18)
ax.tick_params(axis = 'x', labelsize = 14)
ax.tick_params(axis = 'y', labelsize = 14)
ax.figure.set_size_inches(16, 8)
plt.title('觀察"單價元平方公尺"在不同"距離分類(捷運站)"區間的分佈差異', fontsize = 24)
plt.show()
```

```python
# 計算距離最近學校的距離(先轉成平面座標)

gdf_real_estate = gdf_real_estate.to_crs(epsg = 3826)

schools = pd.read_csv('./data/Schools_cleaned.csv', encoding = 'utf-8')
print(schools)

gdf_schools = gpd.GeoDataFrame(schools, geometry = gpd.points_from_xy(schools['twd97_x'], schools['twd97_y']), crs = 'EPSG: 3826')

print(gdf_schools)
print(gdf_schools.shape)

gdf_real_estate['距學校距離'] = gdf_real_estate.geometry.apply(lambda p: gdf_schools.distance(p).min())

gdf_real_estate['距離分類(學校)'] = pd.cut(gdf_real_estate['距學校距離'],
                                    bins = [0, 500, 1000, 1500, 2000],
                                    labels = ['0-500m', '500-1000m', '1000-1500m', '1500-2000m'])
print(gdf_real_estate)

# 統計各種不同學校距離的房屋數量
school_distance = gdf_real_estate.groupby('距離分類(學校)')
print(school_distance.size())

# 算出各種不同學校距離的房屋均價
print(school_distance['單價元平方公尺'].mean())

# 設定學校緩衝區
gdf_schools['buffer_500m(schools)'] = gdf_schools.buffer(500)
print(gdf_schools)

# 建立學校房價列表
result_list_schools = []

for idx, school in gdf_schools.iterrows():
    nearby = gdf_real_estate[gdf_real_estate.geometry.within(school['buffer_500m(schools)'])]
    mean_price = nearby['單價元平方公尺'].mean()
    result_list_schools.append({'學校名': school['地標名稱'], '平均房價': mean_price, '房屋數量': len(nearby)})
df_schools_prices = pd.DataFrame(result_list_schools)
print(result_list_schools)

print(df_schools_prices)

# df_schools_prices.to_csv('./data/學校房價統計表.csv', index = 0, header = 1, encoding = 'utf-8')

# 繪出'距學校距離'直方圖
sns.histplot(gdf_real_estate['距離分類(學校)'])
plt.show()

# 用boxplot觀察'單價元平方公尺'在不同'距離分類(學校)'區間的分佈差異
b = gdf_real_estate.sort_values(by = '距離分類(學校)', ascending = True)
ax = sns.boxplot(x = '距離分類(學校)', y = '單價元平方公尺', data = b)
plt.xlabel('距離分類(學校)', fontsize = 18)
plt.ylabel('單價元平方公尺', fontsize = 18)
ax.tick_params(axis = 'x', labelsize = 14)
ax.tick_params(axis = 'y', labelsize = 14)
ax.figure.set_size_inches(16, 8)
plt.title('觀察"單價元平方公尺"在距離學校不同"距離分類(學校)"區間的分佈差異', fontsize = 24)
plt.show()
```

```python
# 計算距離最近超市的距離(先轉成平面座標)

gdf_real_estate = gdf_real_estate.to_crs(epsg = 3826)

markets = pd.read_csv('./data/Zhonghe_markets_cleaned.csv', encoding = 'utf-8')
print(markets)

gdf_markets = gpd.GeoDataFrame(markets, geometry = gpd.points_from_xy(markets.Lng, markets.Lat), crs = 'EPSG: 4326')
gdf_markets = gdf_markets.to_crs(epsg = 3826)

print(gdf_markets)
print(gdf_markets.shape)

gdf_real_estate['距超市距離'] = gdf_real_estate.geometry.apply(lambda p: gdf_markets.distance(p).min())
gdf_real_estate['距離分類(超市)'] = pd.cut(gdf_real_estate['距超市距離'],
                                     bins = [0, 500, 1000, 1500, 2000],
                                     labels = ['0-500m', '500-1000m', '1000-1500m', '1500-2000m'])
print(gdf_real_estate)

# 統計各種不同超市距離的房屋數量
markets_distance = gdf_real_estate.groupby('距離分類(超市)')
print(markets_distance.size())

# 算出各種不同超市距離的房屋均價
print(markets_distance['單價元平方公尺'].mean())

# 設定超市緩衝區
gdf_markets['buffer_500m(超市)'] = gdf_markets.buffer(500)
print(gdf_markets)

# 建立超市房價列表
result_list_markets = []

for idx, market in gdf_markets.iterrows():
    nearby = gdf_real_estate[gdf_real_estate.geometry.within(market['buffer_500m(超市)'])]
    mean_price = nearby['單價元平方公尺'].mean()
    result_list_markets.append({'超市名': market['name'], '平均房價': mean_price, '房屋數量': len(nearby)})
df_markets_prices = pd.DataFrame(result_list_markets)

print(result_list_markets)
print(df_markets_prices)

# df_markets_prices.to_csv('./data/超市房價統計表.csv', index = 0, header = 1, encoding = 'utf-8')

# 繪出'距超市距離'直方圖
sns.histplot(gdf_real_estate['距離分類(超市)'])
plt.show()

# 用boxplot觀察'單價元平方公尺'在不同'距離分類(超市)'區間的分佈差異
c = gdf_real_estate.sort_values(by = '距離分類(超市)', ascending = True)
ax = sns.boxplot(x = '距離分類(超市)', y = '單價元平方公尺', data = c)

plt.xlabel('距離分類(超市)', fontsize = 18)
plt.ylabel('單價元平方公尺', fontsize = 18)
ax.tick_params(axis = 'x', labelsize = 14)
ax.tick_params(axis = 'y', labelsize = 14)
ax.figure.set_size_inches(16, 8)
plt.title('觀察"單價元平方公尺"在距離超市不同"距離分類(超市)"區間的分佈差異', fontsize = 24)
plt.show()
```

```python
# 計算最近公園之距離(先轉成平面座標)

gdf_real_estate = gdf_real_estate.to_crs(epsg = 3826)

parks = pd.read_csv('./data/park_cleaned.csv', encoding = 'utf-8')
print(parks)

gdf_parks = gpd.GeoDataFrame(parks, geometry = gpd.points_from_xy(parks.Lng, parks.Lat), crs= 'EPSG: 4326')
gdf_parks = gdf_parks.to_crs(epsg = 3826)

print(gdf_parks)
print(gdf_parks.shape)

gdf_real_estate['距公園距離'] = gdf_real_estate.geometry.apply(lambda p: gdf_parks.distance(p).min())
gdf_real_estate['距離分類(公園)'] = pd.cut(gdf_real_estate['距公園距離'],
                                    bins = [0, 500, 1000, 1500, 2000],
                                    labels = ['0-500m', '500-1000m', '1000-1500m', '1500-2000m'])
print(gdf_real_estate)

# 統計各種不同公園距離的房屋數量
parks_distance = gdf_real_estate.groupby('距離分類(公園)')
print(parks_distance.size())

# 算出各種不同公園距離的房屋均價
print(parks_distance['單價元平方公尺'].mean())

# 設定公園緩衝區
gdf_parks['buffer_500m(公園)'] = gdf_parks.buffer(500)
print(gdf_parks)

# 建立公園房價列表
result_list_parks = []

for idx, park in gdf_parks.iterrows():
    nearby = gdf_real_estate[gdf_real_estate.geometry.within(park['buffer_500m(公園)'])]
    mean_prices = nearby['單價元平方公尺'].mean()
    result_list_parks.append({'公園名': park['name'], '平均房價': mean_prices, '房屋數量': len(nearby)})
df_parks_prices = pd.DataFrame(result_list_parks)

print(result_list_parks)
print(df_parks_prices)

# df_parks_prices.to_csv('./data/公園房價統計表.csv', index = 0, header = 1, encoding = 'utf-8')

# 繪出公園距離直方圖
sns.histplot(gdf_real_estate['距離分類(公園)'])
plt.show()

# 用boxplot觀察'單價元平方公尺'在不同'距離分類(公園)'區間的分佈差異
d = gdf_real_estate.sort_values(by = '距離分類(公園)', ascending = True)
ax = sns.boxplot(x = '距離分類(公園)', y = '單價元平方公尺', data = d)

plt.xlabel('距離分類(公園)', fontsize = 18)
plt.ylabel('單價元平方公尺', fontsize = 18)
plt.tick_params(axis = 'x', labelsize = 14)
plt.tick_params(axis = 'y', labelsize = 14)
ax.figure.set_size_inches(16, 8)
plt.title('觀察"單價元平方公尺"在距離超市不同"距離分類(超市)"區間的分佈差異', fontsize = 24)
plt.show()
```

```python
# 多組柱狀圖
import numpy as np
from matplotlib.pyplot import MultipleLocator
import matplotlib.ticker as mticker


df_real_estate = pd.DataFrame(gdf_real_estate)
print(df_real_estate)

distance = ('0-500m', '500-1000m', '1000-1500m', '1500-2000m')
a = tuple(mrt_distance['單價元平方公尺'].mean().to_list())

b = tuple(school_distance['單價元平方公尺'].mean().to_list())

c = tuple(markets_distance['單價元平方公尺'].mean().to_list())

d = tuple(parks_distance['單價元平方公尺'].mean().to_list())

distance_landmarks = {
    '捷運站': a,
    '學校': b,
    '超市': c,
    '公園': d
}

x = np.arange(len(distance))
width = 0.2

multiplier = 0

fig, ax = plt.subplots(layout = 'constrained')
for attribute, landmark in distance_landmarks.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, landmark, width, label = attribute)
    ax.bar_label(rects, padding = 4)
    multiplier += 1

ax.set_ylabel('單價元平方公尺', fontsize = 16)
ax.set_title('不同地標距離分類與單價的關係', fontsize = 20)
ax.set_xticks(x + 6 * width / 4, distance)
ax.figure.set_size_inches(10, 8)
plt.tick_params(axis = 'both', which = 'major', labelsize = 14)

ax.legend(bbox_to_anchor = (1.05, 0), loc = 3, borderaxespad = 0)

plt.show()
```
