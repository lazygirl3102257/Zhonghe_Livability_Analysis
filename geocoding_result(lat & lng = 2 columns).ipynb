```python
import pandas as pd

df = pd.read_csv('./data/地址經緯度.csv', encoding = 'utf-8')
print(df)

df['Lng'] = df['geometry'].map(lambda x: x.split('    ')[0])
# 將索引值重設為原本的樣子
df.reset_index(inplace = True)
# 修改欄位名稱
df.rename(columns = {'index': 'Lat'}, inplace = True)
filtered_data = df.loc[:, ['Lat', 'Lng']]
print(filtered_data)

# 將Lng欄提取出Series並整理前綴字和後綴符號
lng_column = pd.Series(df['Lng'])
tidy_Lng = lng_column.str.replace("'lng': ", "")
tidy_Lng2 = tidy_Lng.str.rstrip('}')
print(tidy_Lng2)

# 將Lat欄提取出Series並整理前綴字
lat_column = pd.Series(df['Lat'])
tidy_Lat = lat_column.str.replace("{'lat': ", "")
print(tidy_Lat)

# 將整理後的Lat和Lng欄拼接起來
geocoding_finished = pd.concat([tidy_Lat, tidy_Lng2], axis = 1, ignore_index = False)
print(geocoding_finished)
```
